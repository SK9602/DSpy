# Step 1. HashTable 클래스 기본적인 메소드 구현하기

- `len(), __len__()` - HashTable에 채워진(사용된) 요소의 갯수를 반환합니다. 
- `__str__()` - HashTable에 채워진 요소(key-value)들을 str type으로 반환합니다. 
- `__repr__()` - HashTable의 모든 내용을 str type으로 반환합니다.

````python
# Given Code
class HashTable:
    def __init__(self):
        self.size = 11
        self.slot = [None] * self.size
        self.data = [None] * self.size
        
    def put(self, key, data):
        hashcode = self.hashfunction(key)

        if self.slot[hashcode] == None:
            self.slot[hashcode] = key
            self.data[hashcode] = data
        elif self.slot[hashcode] == key:
            self.data[hashcode] = data          # update - replace
        else:
            nextslot = self.rehash(hashcode)
            while self.slot[nextslot] is not None and self.slot[nextslot] != key:
                nextslot = self.rehash(nextslot)
            
            if self.slot[nextslot] is None: 
                self.slot[nextslot] = key
                self.data[nextslot] = data
            else:
                self.data[nextslot] = data #replace

    def hashfunction(self, key):
        return key % self.size

    def rehash(self, oldhash):
        return (oldhash + 1) % self.size

    # 목차 4
    def get(self,key):
        startslot = self.hashfunction(key)
        position = startslot
        
        while self.slot[position] != None:
            if self.slot[position] == key:
                return self.data[position]
            else:
                position=self.rehash(position)
                if position == startslot:
                    return None
        return None

    def __getitem__(self,key):
        return self.get(key)

    def __setitem__(self,key,data):
        self.put(key,data)
        
    # Implement From Here
    def __delitem__(self,key):
        pass
    
    def len(self):
        pass
    
    def __len__(self):
        pass
        
    def __str__(self):
        pass
    
    def __repr__(self):
        pass
    
if __name__ == '__main__':
    ht = HashTable()
    keys = [54, 26, 93, 17, 77, 31, 44, 55, 20]
    data = ["cat", "dog", "lion", "ox", "bird", "cow", "goat", "pig", "bear"]
    for key, data in zip(keys, data):  
        ht.put(key, data)

    print('\nSTEP 1: provide basic functions len(), __len__(), __str__(), __repr__()\n')
    print('          len():', ht.len())                   # size of items in dict
    print('      __len__():', ht.__len__())               # size of items in dict
    print('      __str__():', ht.__str__())               # str format of elements used in table
    print('     __repr__():', ht.__repr__())              # str format of all elements in table       
````
